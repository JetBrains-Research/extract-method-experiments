```python
import pandas as pd
```

```python
pos = pd.read_csv('datasets/oct_26_positive.csv', delimiter=';')
neg = pd.read_csv('datasets/oct_26_negative.csv', delimiter=';')
```

```python
pos = pos.drop(columns=pos.columns[0])
neg = neg.drop(columns=neg.columns[0])
```

```python
pos_test = pos.sample(1000)
neg_test = neg.sample(1000*50)
```

```python
pos_train = pos.drop(pos_test.index)
neg_train = neg.drop(neg_test.index).sample(int(1e6))
```
